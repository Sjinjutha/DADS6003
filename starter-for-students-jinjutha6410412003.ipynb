# Cleansing data

```python
# check data type in training data

df_train.dtypes
```

```python
# check missing values in training data

df_train.isnull().sum()
```

```python
n = df_train.shape[0]
print(f'Sample size is {n}')
```

```python
# check distribution of each variable

plt.figure(figsize=[20, 15])

col = list(df_train.columns.values)

for i in range(1,21):
    plt.subplot(7,3,i)
    ax = df_train.iloc[:,i-1].hist(bins=15, density=True, stacked=True, color='teal', alpha=0.6)
    df_train.iloc[:,i-1].plot(kind='density', color='teal')
    plt.title(col[i-1], fontsize=16, fontweight='bold')
```

```python
# mean and median all variables

for i in list(df_train.columns.values):
    print(f'The mean and median of "{i}" are {df_train[i].mean(skipna=True):.2f}, {df_train[i].median(skipna=True):.2f}')
```

use mean to fill in missing value because each variables are approximately normal distribution

```python
train_data = df_train.copy()

col = list(df_train.columns.values)
for i in col:
    train_data[i].fillna(df_train[i].mean(skipna=True), inplace=True)

train_data.isnull().sum()
```

```python
plt.figure(figsize=[20, 15])

col = list(df_train.columns.values)

for i in range(1,21):
    plt.subplot(7,3,i)
    ax = train_data.iloc[:,i-1].hist(bins=15, density=True, stacked=True, color='teal', alpha=0.6)
    train_data.iloc[:,i-1].plot(kind='density', color='teal')
    plt.title(col[i-1], fontsize=16, fontweight='bold')
```

after filled mean in missing value, each variables are still approximately normal distribution

```python
x_train = train_data.loc[:, train_data.columns != 'y']
y_train = train_data['y']
```

```python
# check missing values in testing data

df_test.isnull().sum()
```

```python
test_data = df_test.copy()
test_data.head()
```

```python
# check correlation between variables

import seaborn as sns

plt.subplots(figsize=(25, 20))
sns.heatmap(x_train.corr(), annot=True, cmap="RdYlGn")
plt.show()
```

data need to fo feature scaling to improve the performance and training stability of the model by 'Standardization'

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_scale = scaler.transform(x_train)

x_scale = pd.DataFrame(x_scale, columns=['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13','x14','x15','x16','x17','x18','x19','x20'])
x_scale.head()
```

Also transform in testing data

```python
scaler.fit(test_data)
test_scale = scaler.transform(test_data)

test_scale = pd.DataFrame(test_scale, columns=['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13','x14','x15','x16','x17','x18','x19','x20'])
test_scale.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x_scale, y_train, test_size=0.20, random_state=42)
```

```python
# Ridge regression with built-in cross-validation.

from sklearn.linear_model import RidgeCV

ridge = RidgeCV(alphas=np.logspace(-6, 6, num=5)).fit(X_train, Y_train)
importance = np.abs(ridge.coef_)
feature_names = np.array(X_train.columns.values)
plt.bar(height=importance, x=feature_names)
plt.title("Feature importances via coefficients")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

knn = KNeighborsClassifier()
param_grid = {'n_neighbors':[i for i in range(2, 8)],'weights':['uniform','distance']}
CV = KFold(n_splits=5,random_state=None, shuffle=False)
  
# defining parameter range
grid = GridSearchCV(knn, param_grid, cv=CV, scoring='accuracy', return_train_score=True, verbose=1)
  
# fitting the model for grid search
grid_search = grid.fit(X_train, Y_train)

print(grid_search.best_params_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=7, weights='uniform')
neigh.fit(X_train, Y_train)

predictions1 = neigh.predict(X_test)
print(predictions1)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(neigh, X_train, Y_train, values_format='d' )


from sklearn.metrics import accuracy_score
y_pred = neigh.predict(X_train)
print(f'Accuracy score is {accuracy_score(Y_train, y_pred) *100}%')

print(f'Accuracy score when using with test data is { neigh.score(X_test, Y_test) *100}%')
```
